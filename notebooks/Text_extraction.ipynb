```python
import fitz  # PyMuPDF
import os

# Function to save image bytes to a file
def save_image(image, image_index, output_folder):
    # Prepare image file name
    image_filename = f"image_{image_index}.png"
    image_path = os.path.join(output_folder, image_filename)

    # Write the image bytes to the file
    with open(image_path, "wb") as img_file:
        img_file.write(image["image"])

    print(f"Image saved: {image_path}")

# Function to extract images from a PDF
def extract_images_from_pdf(pdf_path, output_folder):
    # Create the output directory if it doesn't exist
    if not os.path.exists(output_folder):
        os.makedirs(output_folder)

    # Open the PDF file
    pdf = fitz.open(pdf_path)
    image_index = 0

    # Loop through each page
    for page_number in range(len(pdf)):
        page = pdf[page_number]
        # Extract image objects from the page
        image_list = page.get_images(full=True)

        for img in image_list:
            xref = img[0]
            # Get the raw image bytes
            base_image = pdf.extract_image(xref)
            if base_image:
                save_image(base_image, image_index, output_folder)
                image_index += 1

    pdf.close()

# Specify the input PDF and output folder
pdf_path = "/content/2005.14165.pdf"
output_folder = "extracted_images2"

# Extract images from the PDF
extract_images_from_pdf(pdf_path, output_folder)
```

```python
import fitz  # PyMuPDF
import os

# Function to save image bytes to a file
def save_image(image, image_index, output_folder, document_name, page_number):
    # Prepare image file name with document name and page number
    image_filename = f"{document_name}_page{page_number}_image{image_index}.png"
    image_path = os.path.join(output_folder, image_filename)

    # Write the image bytes to the file
    with open(image_path, "wb") as img_file:
        img_file.write(image["image"])

    print(f"Image saved: {image_path}")

# Function to extract images from a single PDF
def extract_images_from_pdf(pdf_path, output_folder):
    # Get the document name without extension for folder and naming
    document_name = os.path.splitext(os.path.basename(pdf_path))[0]
    # Create a dedicated output directory for each document
    document_output_folder = os.path.join(output_folder, document_name)
    if not os.path.exists(document_output_folder):
        os.makedirs(document_output_folder)

    # Open the PDF file
    pdf = fitz.open(pdf_path)
    image_index = 0

    # Loop through each page
    for page_number in range(len(pdf)):
        page = pdf[page_number]
        # Extract image objects from the page
        image_list = page.get_images(full=True)

        for img in image_list:
            xref = img[0]
            # Get the raw image bytes
            base_image = pdf.extract_image(xref)
            if base_image:
                save_image(base_image, image_index, document_output_folder, document_name, page_number + 1)
                image_index += 1

    pdf.close()

# Function to extract images from multiple PDFs
def extract_images_from_multiple_pdfs(pdf_paths, output_folder):
    for pdf_path in pdf_paths:
        extract_images_from_pdf(pdf_path, output_folder)

# Specify the list of input PDFs and output folder
pdf_files = [
    '/content/2306.11695.pdf',
    '/content/1710.05941.pdf',
    '/content/2110.15343.pdf',
    '/content/2310.06825.pdf',
    '/content/2306.17806.pdf',
    '/content/2306.12929.pdf',
    '/content/ICML03-094.pdf',
    '/content/2301.00774.pdf',
    '/content/2005.14165.pdf',
    '/content/Adaptive-mixtures-of-local-experts.pdf',
    '/content/1907.01470.pdf',
    '/content/2305.18290.pdf',
    '/content/2310.20707.pdf',
    '/content/2207.00112.pdf',
    '/content/2211.05102.pdf',
    '/content/2203.02155.pdf',
    '/content/2104.09864.pdf',
    '/content/2307.13304.pdf',
    '/content/2001.08361.pdf',
    '/content/2302.10866.pdf',
    '/content/2305.07185.pdf',
    '/content/2306.07629.pdf',
    '/content/10000000_662098952474184_2584067087619170692_n.pdf',
    '/content/2109.06243.pdf',
    '/content/2301.13688.pdf'
]
output_folder = "extracted_images"

# Extract images from the PDFs
extract_images_from_multiple_pdfs(pdf_files, output_folder)
```
